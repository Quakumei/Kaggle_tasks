В рамках данного кода решается следующая задача. У нас есть страница, на которой лежат ссылки на файлы в виде таблицы, причём файлов много больше сотни. При этом задача будет регулярно повторятся в дальнейшем, так что есть смысл написать конкретно под эту страницу полноценный даунлодер.

```python
import requests
import urllib
from bs4 import BeautifulSoup

url = "http://localhost:8888/downloads/" #адрес страницы для парсинга
page = requests.get(url) #загрузка страницы
soup = BeautifulSoup(page.text,"html.parser") #парсинг страницы в читаемый суп
path = "D:\\jupyter\\downloads\\" #путь для сохранения файлов
url = "http://localhost:8888/" #переопределяем URL для того чтобы нам дальше было проще строить ссылки
```

```python
print(soup.prettify())
#Показываем код сайта в красивом читаемом виде,
#чтобы было проще искать те структуры внутри которых лежат ссылки
```

```python
#Создаём пустой список в который мы будем сохранять ссылки
links = []
#для каждого div найденного в супе
for div in soup.find_all("div"):
    #если div это class
    if div.has_attr("class"):
        #и если в этом классе есть экземпляр request-completed
        if 'request-completed' in div.get('class'):
            #для каждого span в table в div
            for span in div.find('table').find_all('span'):
                #если span это class
                if span.has_attr('class'):
                    #и в нём есть экземпляр identifier_with_uri
                    if 'identifier_with_uri' in span.get('class'):
                        #для каждой ссылки(a href) в span
                        for href in span.find_all("a"):
                            #напечатать название исходя из логики страницы
                            print (href.text[7:])
                            #сгенерировать ссылку опять же по тем правилам которые есть на этой конкретной странице
                            link = url + 'freenet:' + href.get('href')[1:] + "?forcedownload=true"
                            #попробовать
                            try:
                                #скачать файл по ссылке и сохранить по пути обьявленному ранее
                                urllib.request.urlretrieve (link, path + href.text[7:])
                            #если не взлетит
                            except:
                                #сообщить что что-то пошло не так
                                print ('Error!')
                            #добавить ссылку в списочек
                            links.append(link)
#показать сколько ссылок было найдено
print(len(links))
```
